```python
import open3d as o3d
import os
import shutil
from tqdm import tqdm
import glob
```

## Banglore Annotation filter

```python
src_dir = "/media/asad/ADAS_CV/pointclouds/planteye_allscans/merge/banglore"
label_dir = "/media/asad/ADAS_CV/pointclouds/banglore_annotation"
dst_src="../data/raw_data/pointclouds"
dst_labels="../data/raw_data/annotation"
```

```python
for fs in tqdm(os.listdir(label_dir)):
    inp=fs[:-5] + ".ply"
    gtfile = os.path.join(label_dir, fs)
    plyfile = os.path.join(src_dir, inp)
    assert os.path.exists(plyfile), f"Input file {plyfile} does not exists!!"
    shutil.copy(gtfile, os.path.join(dst_labels,fs))
    shutil.copy(plyfile, os.path.join(dst_src,inp))
    
```

## AWS annotaiton

```python
label_files = list()
path_label_files =list()
search_dir = "/media/asad/8800F79D00F79104/ann_pc_aws" 
for (dirpath, dirnames, filenames) in os.walk(search_dir ):
    path_label_files += [os.path.join(dirpath, file) for file in filenames if file.endswith(".zlib")]
    label_files += [file  for file in filenames if file.endswith(".zlib")]
```

```python
ply_files = list()
path_ply_files =list()
search_dir = "/media/asad/8800F79D00F79104/ann_pc_aws" 
for (dirpath, dirnames, filenames) in os.walk(search_dir ):
    path_ply_files += [os.path.join(dirpath, file) for file in filenames if file.endswith(".ply")]
    ply_files += [file  for file in filenames if file.endswith(".ply")]
```

```python
unique_labels=set(label_files)
```

```python
for ul in tqdm(unique_labels):
    s_labels=[s for s in path_label_files if ul in s]
    s_labels.sort(key=lambda x: os.path.getmtime(x),reverse=True)
    assert len(s_labels), "No labels!!"
    l=s_labels[0]
    fl=l.split("/")[-1].split(".")[0]
    match_inp=[s for s in path_ply_files if fl in s]
    i=match_inp[0]
    shutil.copy(i, os.path.join(dst_src,i.split("/")[-1]))
    shutil.copy(l, os.path.join(dst_labels,l.split("/")[-1]))
    #print(l)
    #print(i)
    #break
```
