```python
import sagemaker
sess=sagemaker.Session(default_bucket="sagemaker-us-east-1-470086202700")
role= sagemaker.get_execution_role()
account =sess.boto_session.client('sts').get_caller_identity()['Account']
region= sess.boto_session.region_name
role
```

```python
#Name of repo
repo_name="dyco3d"
base_job_name="dyco3d"
#train_input_path="s3://sagemaker-us-east-1-470086202700/3d-segmentation/test_data/data/"
train_input_path="s3://sagemaker-us-east-1-470086202700/3d-segmentation/dyco3d-2021-data/"
```

```python
image_name  = '{}.dkr.ecr.{}.amazonaws.com/{}:latest'.format(account, region, repo_name)
output_path="s3://sagemaker-us-east-1-470086202700/3d-segmentation/dyco3d-model/"
image_name
```

```python
train_instance_type ="ml.p3.8xlarge"
```

```python
estimator = sagemaker.estimator.Estimator(
                       image_uri=image_name,
                       base_job_name=base_job_name,
                       role=role, 
                       instance_count=1, 
                       instance_type=train_instance_type,
                       output_path=output_path,
                       sagemaker_session=sess,
   #                    hyperparameters={
   #                           'epochs': 20,
   #                           'batch-size': 130,
   #                           'lr': 0.05}
                        )
```

## Start Training

```python
estimator.fit({'training': train_input_path})
```

```python
## Hyperparameter tuning
```

```python
# !aws sagemaker stop-hyper-parameter-tuning-job --hyper-parameter-tuning-job-name 
```

```python
from sagemaker.tuner import IntegerParameter, CategoricalParameter, ContinuousParameter, HyperparameterTuner

hyperparameter_ranges = {
   # 'lr': ContinuousParameter(0.0001, 0.01, scaling_type='Logarithmic'), 
    'cluster_radius': ContinuousParameter(0.028, 0.05),
    'cluster_meanActive':    IntegerParameter(36, 52),
    'cluster_shift_meanActive':   IntegerParameter(240, 310),
    'cluster_npoint_thre':       IntegerParameter(10, 25)
}
```

```python
hyperparameter_ranges
```

```python
objective_metric_name = 'val_acc'
objective_type = 'Maximize'
metric_definitions = [{'Name':'val_acc', 'Regex':'Best AP:([0-9\\.]+)'}]

tuner = HyperparameterTuner(estimator,
                            objective_metric_name,
                            hyperparameter_ranges,
                            metric_definitions,
                            max_jobs=50,
                            max_parallel_jobs=2,
                            objective_type=objective_type)
```

```python
tuner.fit({'training': train_input_path})
```
